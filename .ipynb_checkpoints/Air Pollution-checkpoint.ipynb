```python
import requests
import json
import pandas as pd
```

```python
my_api_key = '31c4a0771a0556ec808b4c09c73e965e'
start='1627776000'
end='1630454400'
lat_long_list = [
    '10.3886, 106.6602', #Gò Công, Tiền Giang
    '11.3254, 106.6602', #Phú Giáo, Bình Dương
    '11.7512, 106.6602', #Bình Long, Bình Phước
    '21.9661, 106.6602', #Văn Lãng, Lạng Sơn
    '17.1540, 106.6602', #Lệ Thủy, Quảng Bình
    '10.8231, 106.6602' #Tân Bình, TP.HCM
]
```

```python
aqi_provinces = []
for ll in lat_long_list:
    latlong = ll.split(", ")
    lat = latlong[0]
    long = latlong[1]
    url = f'http://api.openweathermap.org/data/2.5/air_pollution/history?lat={lat}&lon={lon}&start={start}&end={end}&appid={my_api_key}'
    json_str = requests.get(url).text
    json_pydata = json.loads(json_str)
    aqi_provinces.append(json_pydata)
```

```python
def split(list):
    new = {}
    new['co']=list['components']['co']
    new['no']=list['components']['no']
    new['no2']=list['components']['no2']
    new['o3']=list['components']['o3']
    new['so2']=list['components']['so2']
    new['pm2_5']=list['components']['pm2_5']
    new['pm10']=list['components']['pm10']
    new['nh3']=list['components']['nh3']
    new['aqi']=list['main']['aqi']
    return new
    
    
```

```python
def remake(data):
    new=[]
    for i in data['list']:
        new.append(split(i))
    return new
    
```

```python
data = []
for prov in aqi_provinces:
    data += remake(prov)
len(data)
```
