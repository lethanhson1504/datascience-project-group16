```python
def split(list):
    new={}
    new['dt']=list['dt']
    new['co']=list['components']['co']
    new['no']=list['components']['no']
    new['no2']=list['components']['no2']
    new['o3']=list['components']['o3']
    new['so2']=list['components']['so2']
    new['pm2_5']=list['components']['pm2_5']
    new['pm10']=list['components']['pm10']
    new['nh3']=list['components']['nh3']
    new['aqi']=list['main']['aqi']
    return new
    
    
```

```python
def remake(data):
    new=[]
    for i in data['list']:
        new.append(split(i))
    return new
    
```

```python
data=remake(json_pydata)
data
```
